```python
%%manim -qm -v WARNING fluid

from manim import *
import random

config.media_embed = True

class fluid(ThreeDScene):
    particle_count = 20
    xyzrange = 5
    
    def construct(self):
        axes = ThreeDAxes()
        self.add(axes)
        particles = self.gen_particles()
        for particle in particles:
            self.add(particle)
        self.set_camera_orientation(phi = 75 * DEGREES, theta = 30 * DEGREES)
        self.begin_ambient_camera_rotation(rate=0.1)
        self.wait()
        self.stop_ambient_camera_rotation()
        self.move_camera(phi = 75 * DEGREES, theta = 30 * DEGREES)
        self.wait()
        
    def gen_particles(self):
        particles = []
        for i in range(self.particle_count):
            x = random.randint(-self.xyzrange, self.xyzrange)
            y = random.randint(-self.xyzrange, self.xyzrange)
            z = random.randint(-self.xyzrange, self.xyzrange)
            p = Dot3D([x, y, z])
            particles.append(p)
        return particles

# Fluid at home (i'll get back to it later)
        
```
